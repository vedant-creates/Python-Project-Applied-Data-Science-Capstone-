```python
import pandas as pd
import requests
import io

# Download and read the `spacex_launch_geo.csv`
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
response = requests.get(URL)
response.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)
spacex_csv_file = io.BytesIO(response.content)
spacex_df = pd.read_csv(spacex_csv_file)
```

Now, you can take a look at what are the coordinates for each site.


```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

```python
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
proximity_data = {
    'Launch Site': ['CCAFS SLC-40', 'CCAFS LC-40', 'KSC LC-39A', 'VAFB SLC-4E'],
    'Launch Site Lat': [28.563197, 28.562302, 28.573255, 34.632834],
    'Launch Site Long': [-80.576820, -80.577356, -80.646895, -120.610745],
    'Coastline Lat': [28.56357, 28.56249, 28.58136, 34.63342],
    'Coastline Long': [-80.56799, -80.56758, -80.64153, -120.6279],
    'Highway Lat': [28.56474, 28.56295, 28.57619, 34.63795],
    'Highway Long': [-80.57092, -80.57067, -80.64553, -120.62097],
    'Railway Lat': [28.57148, 28.57148, 28.57398, 34.63457],
    'Railway Long': [-80.58534, -80.58534, -80.65416, -120.62433],
    'City Lat': [28.53392, 28.53392, 28.53392, 34.95124],
    'City Long': [-81.35633, -81.35633, -81.35633, -120.44181]
}
proximity_df = pd.DataFrame(proximity_data)

proximity_df['Distance to Coastline'] = proximity_df.apply(lambda row: calculate_distance(row['Launch Site Lat'], row['Launch Site Long'], row['Coastline Lat'], row['Coastline Long']), axis=1)
proximity_df['Distance to Highway'] = proximity_df.apply(lambda row: calculate_distance(row['Launch Site Lat'], row['Launch Site Long'], row['Highway Lat'], row['Highway Long']), axis=1)
proximity_df['Distance to Railway'] = proximity_df.apply(lambda row: calculate_distance(row['Launch Site Lat'], row['Launch Site Long'], row['Railway Lat'], row['Railway Long']), axis=1)
proximity_df['Distance to City'] = proximity_df.apply(lambda row: calculate_distance(row['Launch Site Lat'], row['Launch Site Long'], row['City Lat'], row['City Long']), axis=1)
```

```python

launch_site_group = folium.FeatureGroup(name="Launch Sites")
coastline_group = folium.FeatureGroup(name="Coastlines")
highway_group = folium.FeatureGroup(name="Highways")
railway_group = folium.FeatureGroup(name="Railways")
city_group = folium.FeatureGroup(name="Cities")

for index, row in proximity_df.iterrows():
    # Add markers to respective FeatureGroups
    folium.Marker([row['Launch Site Lat'], row['Launch Site Long']],
                 popup=row['Launch Site'],
                 icon=folium.Icon(color='orange')).add_to(launch_site_group)

    distance_to_coastline = calculate_distance(row['Launch Site Lat'], row['Launch Site Long'],
                                                row['Coastline Lat'], row['Coastline Long'])
    folium.Marker([row['Coastline Lat'], row['Coastline Long']],
                 popup=f'Coastline ({distance_to_coastline:.2f} km)',
                 icon=folium.Icon(color='blue')).add_to(coastline_group)

    distance_to_highway = calculate_distance(row['Launch Site Lat'], row['Launch Site Long'],
                                              row['Highway Lat'], row['Highway Long'])
    folium.Marker([row['Highway Lat'], row['Highway Long']],
                 popup=f'Highway ({distance_to_highway:.2f} km)',
                 icon=folium.Icon(color='green')).add_to(highway_group)

    distance_to_railway = calculate_distance(row['Launch Site Lat'], row['Launch Site Long'],
                                              row['Railway Lat'], row['Railway Long'])
    folium.Marker([row['Railway Lat'], row['Railway Long']],
                 popup=f'Railway ({distance_to_railway:.2f} km)',
                 icon=folium.Icon(color='black')).add_to(railway_group)

    distance_to_city = calculate_distance(row['Launch Site Lat'], row['Launch Site Long'],
                                            row['City Lat'], row['City Long'])
    folium.Marker([row['City Lat'], row['City Long']],
                 popup=f'City ({distance_to_city:.2f} km)',
                 icon=folium.Icon(color='red')).add_to(city_group)

# Add FeatureGroups to the map
site_map.add_child(launch_site_group)
site_map.add_child(coastline_group)
site_map.add_child(highway_group)
site_map.add_child(railway_group)
site_map.add_child(city_group)


# Add LayerControl to toggle visibility of layers
folium.LayerControl().add_to(site_map)
site_map
```

```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
# lines=folium.PolyLine(locations=coordinates, weight=1)

for index, row in proximity_df.iterrows():


  coastline_polyline = folium.PolyLine(
        locations=[
            [row['Launch Site Lat'], row['Launch Site Long']],
            [row['Coastline Lat'], row['Coastline Long']],
        ],
        color='blue',
        weight=1,
    ).add_to(coastline_group)
  folium.Popup(f"Distance to Coastline: {row['Distance to Coastline']:.2f} km",
                 parse_html=True).add_to(coastline_polyline)

    # Create and add a PolyLine for highway
  highway_polyline = folium.PolyLine(
        locations=[
            [row['Launch Site Lat'], row['Launch Site Long']],
            [row['Highway Lat'], row['Highway Long']],
        ],
        color='green',
        weight=1,
    ).add_to(highway_group)
  folium.Popup(f"Distance to Highway: {row['Distance to Highway']:.2f} km",
                 parse_html=True).add_to(highway_polyline)

    # Create and add a PolyLine for railway
  railway_polyline = folium.PolyLine(
        locations=[
            [row['Launch Site Lat'], row['Launch Site Long']],
            [row['Railway Lat'], row['Railway Long']],
        ],
        color='black',
        weight=1,
    ).add_to(railway_group)
  folium.Popup(f"Distance to Railway: {row['Distance to Railway']:.2f} km",
                 parse_html=True).add_to(railway_polyline)


    # Create and add a PolyLine for city
  city_polyline = folium.PolyLine(
        locations=[
            [row['Launch Site Lat'], row['Launch Site Long']],
            [row['City Lat'], row['City Long']],
        ],
        color='red',
        weight=1,
    ).add_to(city_group)
  folium.Popup(f"Distance to City: {row['Distance to City']:.2f} km",
                 parse_html=True).add_to(city_polyline)

# Add FeatureGroups to the map
  site_map.add_child(launch_site_group)
  site_map.add_child(coastline_group)
  site_map.add_child(highway_group)
  site_map.add_child(railway_group)
  site_map.add_child(city_group)

# Display the map
site_map
```
